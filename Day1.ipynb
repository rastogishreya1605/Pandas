6. Select a row using loc

```python
import pandas as pd
a=pd.DataFrame({
    "Name":["Shreya","Satyam","Divya","Tiku"],
    "Marks":[90,95,85,86]
})
print(a.loc[:])
```

7.  Use loc to select the row with index 0 

```python
import pandas as pd
a=pd.DataFrame({
    "Name":["Shreya","Satyam","Divya","Tiku"],
    "Marks":[90,95,85,86]
})
print(a.loc[0:])
```

8. Use iloc to select the first row

```python
import pandas as pd
a=pd.DataFrame({
    "Name":["Shreya","Satyam","Divya","Tiku"],
    "Marks":[90,95,85,86]
})
print(a.iloc[0])
```

9. Use loc to get only the Marks of index 2

```python
import pandas as pd
a=pd.DataFrame({
    "Name":["Shreya","Satyam","Divya","Tiku"],
    "Marks":[90,95,85,86]
})
print(a.loc[2, "Marks"])
```

10.  Use iloc to get marks using row and column position.

```python
import pandas as pd
data = {
    "Name": ["Shreya", "Satyam", "Divya"],
    "Age": [21, 22, 23],
    "Marks": [80, 75, 90]
}
df = pd.DataFrame(data)
print(df)
print(df.iloc[1, 2])
print(df.iloc[0, 2])
print(df.iloc[2, 2])
```

11. Select first two rows using both loc and iloc and observe the difference.

```python
import pandas as pd
data = {
    "Name": ["Shreya", "Satyam", "Divya"],
    "Age": [21, 22, 23],
    "Marks": [80, 75, 90]
}
df = pd.DataFrame(data)
print(df.iloc[0:2])
print(df.loc[0:1])
```

12. Convert numpy array into DataFrame-

```python
import numpy as np
import pandas as pd
# NumPy array 
arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])
# NumPy array convert into DataFrame 
df = pd.DataFrame(arr, columns=["A", "B", "C"], index=["Row1", "Row2", "Row3"])
print(arr) # numpy array
print(df)  # NumPy array convert into DataFrame 
```
