1. Load the CSV dataset

3. Filter students scoring above 80

```python
import pandas as pd

data = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Shreya"],
    "Marks": [75, 85, 90],
    "Subjects": ["maths","English","Science"]
})
data[data["Marks"] > 80]
```

4. Sort students by marks (highest first)

```python
data.sort_values(by="Marks", ascending=False)
```

5. Count students per subject

```python
data["Subjects"].value_counts()
```

6. Fill missing marks with mean

```python
import pandas as pd
data = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Shreya"],
    "Marks": [80, None, 90]
})
data["Marks"] = data["Marks"].fillna(data["Marks"].mean())
print(data)
```

7. Group data by subject & average marks

```python
import pandas as pd

data = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Shreya"],
    "Subject": ["Math", "Math", "Science"],
    "Marks": [80, 90, 85]
})
print(data.groupby("Subject")["Marks"].mean())
```

8. Save cleaned dataset to CSV

```python
data.to_csv("clean_students.csv", index=False)
```
