1. Rename Marks → Score

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B"],
    "Marks": [50, 80]
})

df = df.rename(columns={"Marks": "Score"})
print(df)
```

2. Drop unnecessary column

```python
df = pd.DataFrame({
    "Name": ["A", "B"],
    "Score": [50, 80],
    "Extra": ["X", "Y"]
})

df = df.drop(columns=["Extra"])
print(df)
```

3. Convert Score to integer

```python
df = pd.DataFrame({
    "Score": ["50", "80"]
})

df["Score"] = df["Score"].astype(int)
print(df.dtypes)
```

4. Convert Class to string

```python
df = pd.DataFrame({
    "Class": [10, 12]
})

df["Class"] = df["Class"].astype(str)
print(df.dtypes)
```

5. Create Result (Pass/Fail)

```python
df = pd.DataFrame({
    "Score": [40, 75]
})

df["Result"] = df["Score"].apply(lambda x: "Pass" if x >= 50 else "Fail")
print(df)
```

6. Create Grade column

```python
df = pd.DataFrame({
    "Score": [45, 65, 85]
})

def grade(x):
    if x >= 80:
        return "A"
    elif x >= 60:
        return "B"
    else:
        return "C"

df["Grade"] = df["Score"].apply(grade)
print(df)
```

7. Create Performance using pd.cut()

```python
df = pd.DataFrame({
    "Score": [30, 60, 90]
})

df["Performance"] = pd.cut(
    df["Score"],
    bins=[0, 50, 75, 100],
    labels=["Low", "Medium", "High"]
)

print(df)
```

8. Count students per performance

```python
df = pd.DataFrame({
    "Performance": ["Low", "High", "Medium", "High"]
})

print(df["Performance"].value_counts())
```

9. Subject-wise summary using groupby()

```python
df = pd.DataFrame({
    "Subject": ["Math", "Math", "Science"],
    "Score": [60, 80, 70]
})

print(df.groupby("Subject")["Score"].mean())
```

10. Sort DataFrame using sort_index()

```python
df = pd.DataFrame({
    "Name": ["A", "B", "C"]
}, index=[2, 0, 1])

df = df.sort_index()
print(df)
```

11. Save dataset

```python
df = pd.DataFrame({
    "Name": ["A", "B"],
    "Score": [50, 80]
})

df.to_csv("data.csv", index=False)
```
